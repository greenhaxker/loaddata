```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
od = pd.read_csv('opendata_ตัวอย่างข้อมูล_เงินฝากV2.csv', encoding='cp874')

od
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert 'sex' column to numerical values
od['sex'] = od['sex'].map({'ชาย': 1, 'หญิง': 2,'(null)': 3})

# Convert 'province' column to numerical values using label encoding
label_encoder = LabelEncoder()
od['province'] = label_encoder.fit_transform(od['province'])

od = od.dropna()

od
```

```python
od = od[od['age'] <= 100]
od = od[od['DEP_TOTAL'] >= 0.01]

od
```

```python

%matplotlib widget
import plotly.express as px

# Select all columns except 'seqnum' for clustering
data_for_clustering = od.drop(columns=['seqnum'])



# Save the clustered data to a new CSV file
od.to_csv('clustered_opendata.csv', index=False)

fig = px.scatter_3d(od, x='age', y='DEP_TOTAL', z='province', color='sex',
                    title='3D Clusters', labels={'age': 'AGE', 'DEP_TOTAL': 'DEP_TOTAL', 'province': 'Province', 'sex': 'Sex'})

fig.show()
```

```python
fig = px.violin(od, x='seqnum', y='DEP_TOTAL', title='Violin Plot of DEP_TOTAL vs seqnum', labels={'seqnum': 'seqnum', 'DEP_TOTAL': 'DEP_TOTAL'})
fig.show()
```
